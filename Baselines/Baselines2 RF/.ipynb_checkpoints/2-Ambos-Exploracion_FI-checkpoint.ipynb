```python
from sklearn.metrics import confusion_matrix, f1_score

y_true = [1, 1, 0, 0]
y_pred = [1, 0, 1, 0]   

tn, fp, fn, tp = confusion_matrix(y_true, y_pred, labels=[0, 1]).ravel()
print(tn, fp, fn, tp)  # 1 1 1 1
print(confusion_matrix(y_true, y_pred, labels=[0, 1]).ravel())
```

```python
r = tp/(tp+fn)
p = tp/(tp+fp)

f1 = 2*r*p/(r+p)

print(f1)
```
